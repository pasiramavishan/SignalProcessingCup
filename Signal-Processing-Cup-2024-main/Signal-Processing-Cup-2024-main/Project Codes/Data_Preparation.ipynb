## Combining Audio Files

```python
import librosa
import soundfile as sf
```

```python
# Intput directory to save the combined files
enrollment_dir = os.path.join('ROBOVOX_SP_CUP_2024', 'data', 'single-channel', 'enrollment')
speaker_folders = os.listdir(enrollment_dir)

# Output directory to save the combined files
output_dir = os.path.join('ROBOVOX_SP_CUP_2024', 'data', 'single-channel', 'combined-audio')

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)
```

```python
# Number of files to combine for each speaker
num_files_to_combine = 3

# Iterate over each speaker's folder
for speaker_folder in speaker_folders:
    speaker_folder_path = os.path.join(enrollment_dir, speaker_folder)

    # List the first num_files_to_combine WAV files in the speaker's folder
    wav_files = [filename for filename in os.listdir(speaker_folder_path)]

    # Combine all WAV files into a single long file
    combined_audio = []
    for wav_file in wav_files:
        wav_file_path = os.path.join(speaker_folder_path, wav_file)
        audio, sr = librosa.load(wav_file_path, sr=None)
        combined_audio.extend(audio)

    # Save the combined audio file
    output_file_path = os.path.join(output_dir, f"{speaker_folder}.wav")
    sf.write(output_file_path, combined_audio, sr)

print("Combination complete. Combined files saved in:", output_dir)
```

## Removing Silent Parts

```python
from pydub import AudioSegment
from pydub.silence import split_on_silence
```

```python
combined_dir = os.path.join('ROBOVOX_SP_CUP_2024', 'data', 'single-channel', 'combined-audio')
speakers = os.listdir(combined_dir)

dataset_dir = os.path.join('ROBOVOX_SP_CUP_2024', 'data', 'single-channel', 'dataset')
```

```python
def _split_audio(file_path, output_path, silence_thresh=-32):
    sound_file = AudioSegment.from_file(file_path)
    # Normalization
    sound_file = sound_file.apply_gain(-sound_file.max_dBFS)
    
    audio_chunks = split_on_silence(sound_file, 
                                    min_silence_len=500,
                                    silence_thresh=silence_thresh,
                                    keep_silence=False)

    # Concatenate the audio segments
    combined_audio = sum(audio_chunks)
    basename = os.path.basename(file_path)
    output_path = os.path.join(output_path, basename)
    
    # Export the combined audio to a new file
    combined_audio.export(output_path, format="wav")

    # Get the full path
    full_path = os.path.abspath(output_path)

    return full_path
```

```python
for audio_file in speakers:
    file_path = os.path.join(combined_dir, audio_file)
    _split_audio(file_path, dataset_dir, silence_thresh = -32)    
```
